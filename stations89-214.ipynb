```python
def plot_ts(data, columns, title):
    plt.figure(figsize=(15, 10))
    
    # Create subplots for each column
    for i, col in enumerate(columns, 1):
        plt.subplot(4, 2, i)
        plt.plot(data['Date'], data[col], label=col)
        plt.title(f'{col} Time Series')
        plt.xlabel('Date')
        plt.ylabel(col)
        plt.grid(True)
    plt.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.90)
    plt.show()
```

```python
plot_ts(st89_data, ['ETo (mm)', 'Precip (mm)', 'Sol Rad (W/sq.m)', 'Air Temp (C)', 'Vap Pres (kPa)', 'Rel Hum (%)', 'Dew Point (C)', 'Soil Temp (C)'], 'Time Series Plot for st89 (1992-2012)')
plot_ts(st214_data, ['ETo (mm)', 'Precip (mm)', 'Sol Rad (W/sq.m)', 'Air Temp (C)', 'Vap Pres (kPa)', 'Rel Hum (%)', 'Dew Point (C)', 'Soil Temp (C)'], 'Time Series Plot for st214 (2013-2022)')
```

```python
# Define a function to plot histograms for given columns
def plot_hist(data, columns, dataset_name):
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(columns, 1):
        plt.subplot(4, 2, i)
        plt.hist(data[col].dropna(), bins=50, color='skyblue', edgecolor='black')
        plt.title(f'{col} Distribution ({dataset_name})')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot histograms for the first dataset (st89)
plot_hist(st89_data, ['ETo (mm)', 'Precip (mm)', 'Sol Rad (W/sq.m)', 'Air Temp (C)', 'Vap Pres (kPa)', 'Rel Hum (%)', 'Dew Point (C)', 'Soil Temp (C)'], 'st89 (1992-2012)')

# Plot histograms for the second dataset (st214)
plot_hist(st214_data, ['ETo (mm)', 'Precip (mm)', 'Sol Rad (W/sq.m)', 'Air Temp (C)', 'Vap Pres (kPa)', 'Rel Hum (%)', 'Dew Point (C)', 'Soil Temp (C)'], 'st214 (2013-2022)')
```

```python
def wind_rose_plot_with_consistent_colors(data, title):
    # Set data
    wind_data = data[['Wind Speed (m/s)', 'Wind Dir (0-360)']].dropna()
    wind_speed = wind_data['Wind Speed (m/s)'].values
    wind_dir = wind_data['Wind Dir (0-360)'].values

    # Set data for the wind rose
    N = 8
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
    radii = np.histogram(wind_dir, bins=np.linspace(0.0, 360, N+1), weights=wind_speed)[0]
    width = pi / 4 * np.ones(N)
    directions = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    colors = plt.cm.jet(np.linspace(0, 1, N))
    
    # Create the wind rose plot
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(1, 1, 1, projection='polar')
    bars = ax.bar(theta, radii, width=width, bottom=0.0, align='edge', color=colors, alpha=0.6)
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_xticks(theta)
    ax.set_xticklabels(directions)
    ax.set_title(title)
    
    plt.show()

# Create wind rose with consistent colors for the first dataset (st89)
wind_rose_plot_with_consistent_colors(st89_data, 'Wind Rose with Consistent Colors for st89 (1992-2012)')
wind_rose_plot_with_consistent_colors(st214_data, 'Wind Rose with Directions for st214 (2013-2022)')
```

```python
def wind_rose_plot_with_percentage(data, title):
    # Set data
    wind_data = data[['Wind Speed (m/s)', 'Wind Dir (0-360)']].dropna()
    wind_speed = wind_data['Wind Speed (m/s)'].values
    wind_dir = wind_data['Wind Dir (0-360)'].values

    # Set data for the wind rose
    N = 8
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
    radii = np.histogram(wind_dir, bins=np.linspace(0.0, 360, N+1))[0]  # Count frequencies
    radii_percentage = radii / sum(radii) * 100  # Convert to percentage
    width = pi / 4 * np.ones(N)
    directions = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    colors = plt.cm.jet(np.linspace(0, 1, N))
    
    # Create the wind rose plot
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(1, 1, 1, projection='polar')
    bars = ax.bar(theta, radii_percentage, width=width, bottom=0.0, align='edge', color=colors, alpha=0.6)
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_xticks(theta)
    ax.set_xticklabels(directions)
    ax.set_title(title)
    
    plt.show()

# Create wind rose with percentage frequencies for the first dataset (st89)
wind_rose_plot_with_percentage(st89_data, 'Wind Rose with Percentage Frequencies for st89 (1992-2012)')
```
